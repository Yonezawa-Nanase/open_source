```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')
iris
```

```python
for columun in iris.columns:
    plt.figure()
    for specie in iris['species'].unique():
        plt.hist(iris[iris.species == specie][columun], bins = 10,alpha = 0.7, label = specie)
    plt.title(columun)
    plt.legend()
```

```python
plt.figure()
for specie in iris['species'].unique():
    plt.scatter(iris[iris.species == specie]['sepal_length'], iris[iris.species == specie]['sepal_width'])
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.figure()
for specie in iris['species'].unique():
    plt.scatter(iris[iris.species == specie]['petal_length'], iris[iris.species == specie]['petal_width'])
plt.xlabel('petal_length')
plt.ylabel('petal_width')
```
