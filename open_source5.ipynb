```python
import folium
import numpy as np
import pandas as pd
kinki = pd.read_csv('kinki_pop.csv')
kinki['人口'] = kinki['人口'].str.replace(',', '').astype('int')
denki = pd.read_csv('電気自動車充電スタンド.csv', encoding='cp932',skiprows=1)
```

```python
# 地図生成

folium_map = folium.Map(location=[kinki['緯度'].mean(), kinki['経度'].mean()])

# マーカープロット
for i,row in kinki.iterrows():
    folium.CircleMarker(
        location=[row['緯度'], row['経度']],
        popup=row['府県庁所在地'],
        radius=row['人口']/20000,
        color='#ff0000',
    fill_color='#ff0000'
    ).add_to(folium_map)

# 地図表示
folium_map
```

```python
# 地図生成

folium_map = folium.Map(location=[denki['緯度'].mean(), denki['経度'].mean()])

# マーカープロット
for i,row in denki.iterrows():
    folium.Marker(
        location=[row['緯度'], row['経度']],
        popup=row['スタンド名'],
    ).add_to(folium_map)

# 地図表示
folium_map
```
